```python
import pandas as pd
from os import environ
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
load_dotenv()
```

```python
engine = create_engine(f"postgresql://{environ['POSTGRES_USER']}:{environ['POSTGRES_PASSWORD']}@localhost:{environ['POSTGRES_PORT']}/{environ['POSTGRES_DB']}")
```

```python
df = pd.read_sql_query("""
SELECT
    url, origin, neighborhood, zone, state_acronym, business_type, listing_type,
    usable_area, floors, unit_types, bedrooms, bathrooms, suites, parking_spaces,
    amenities, lat, lon, price, condo_fee, created_date
FROM properties LIMIT 10
""", engine).set_index(["url", "origin"])

df
```
